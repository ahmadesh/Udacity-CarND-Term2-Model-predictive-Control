```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
data = pd.read_csv("./xcode/Debug/output.csv")
```

```python
print(data.head())
```

```python
plt.figure(figsize=(10, 9))
fig, ax = plt.subplots(figsize=(15, 5))

plt.subplot(1,2,1)
plt.plot(data['cte'][1:45], 'k-o', linewidth=3.0)
plt.ylim((-3.5,0))
plt.xlabel('Step number', size=18)
plt.ylabel('CTE', size= 18)

plt.subplot(1,2,2)
plt.plot(data['epsi'][1:50], 'k-o', linewidth=3.0)
plt.xlabel('Step number', size=18)
plt.ylabel('EPSI', size= 18)
```

```python
plt.figure(figsize=(18, 18))
fig, ax = plt.subplots(figsize=(15, 10))
plt.subplot(2,2,1)
plt.plot(data['delta'][1:45], 'k-o', linewidth=3.0)
plt.xlabel('Step number', size=18)
plt.ylabel('Steering angle', size= 18)

plt.subplot(2,2,2)
plt.plot(data['throttle'][1:45], 'k-o', linewidth=3.0)
plt.xlabel('Step number', size=18)
plt.ylabel('Throttle', size= 18)

plt.subplot(2,2,3)
plt.plot(data['v'][1:45], 'k-o', linewidth=3.0)
plt.xlabel('Step number', size=18)
plt.ylabel('Velocity', size= 18)

```
